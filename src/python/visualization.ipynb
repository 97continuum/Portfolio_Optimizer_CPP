```python
# Import Modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Data
results_folder = r"/Users/talhajamal/Desktop/Code/Portfolio_Optimizer_CPP/data/results/"
returns = pd.read_csv(results_folder+"PortfolioReturns.csv", header=0)
cov = pd.read_csv(results_folder+"Covariances.csv", header=0)
```

```python
returns
```

```python
avg_return_per_target = returns.mean(axis = 1)
avg_cov_per_target = cov.mean(axis = 1)
SR = (returns.mean(axis=1))/(returns.std(axis=1))
target = [0.005*(i) for i in range(1, 21)]

portfolio_metrics = pd.DataFrame(
    {
        "Target": target,
        "Average Return": avg_return_per_target,
        "Average Covariance": avg_cov_per_target,
        "Average Sharpe Ratio": SR
    }
)
portfolio_metrics
```

```python
percentage_diff_return = (portfolio_metrics['Target'] - portfolio_metrics['Average Return']) / portfolio_metrics['Target']
percentage_diff_return
```

```python
# Plot every 5th row on the same plot
plt.figure(figsize=(15, 10))

for idx in range(0, 20):
    if idx in [0, 4, 9, 14, 19]:
        row = returns.iloc[idx]
        plt.plot(row.index[1:], row.values[1:], marker='o', label=f'Target Return: {row["Target Returns"]:.3f}', alpha=0.7)

plt.title('Returns for Selected Target Returns Across Backtest Periods')
plt.xlabel('Backtest Period')
plt.ylabel('Return')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(loc='upper left', fontsize='small')  # Place legend outside the plot
plt.tight_layout()  # Adjust layout to fit everything

plt.show()
```
