```python
# Import Modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Data
results_folder = r"/Users/talhajamal/Desktop/Code/Portfolio_Optimizer_CPP/data/results/"
returns = pd.read_csv(results_folder+"PortfolioReturns.csv", header=0)
cov = pd.read_csv(results_folder+"Covariances.csv", header=0)
```

```python
returns
```

```python
avg_return_per_target = returns.mean(axis = 1)
avg_cov_per_target = cov.mean(axis = 1)
SR = (returns.mean(axis=1))/(returns.std(axis=1))
target = [0.005*(i) for i in range(1, 21)]

portfolio_metrics = pd.DataFrame(
    {
        "Target": target,
        "Average Return": avg_return_per_target,
        "Average Covariance": avg_cov_per_target,
        "Average Sharpe Ratio": SR
    }
)
portfolio_metrics
```

```python
percentage_diff_return = (portfolio_metrics['Target'] - portfolio_metrics['Average Return']) / portfolio_metrics['Target']
percentage_diff_return
```

```python
# Plot every 5th row on the same plot
plt.figure(figsize=(15, 8))

for idx in range(0, 20):
    if idx in [0, 4, 9, 14, 19]:
        row = returns.iloc[idx]
        plt.plot(row.index[1:], row.values[1:], marker='o', label=f'Target Return: {row["Target Returns"]:.3f}', alpha=0.7)

plt.title('Returns for Selected Target Returns Across Backtest Periods')
plt.xlabel('Backtest Period')
plt.ylabel('Return')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(loc='upper left', fontsize='small')  # Place legend outside the plot
plt.tight_layout()  # Adjust layout to fit everything

plt.show()
```

```python
cumulative_returns = (returns.drop(columns="Target Returns") + 1).cumprod()
cumulative_returns
```

```python
# Plot every 5th row on the same plot
plt.figure(figsize=(12, 8))

for index, row in cumulative_returns.iterrows():
    plt.plot(row, label=f'Target Return: {returns.iloc[index]["Target Returns"]:.3f}')

plt.title('Cumulative Return of $1 Invested in Each Target Return Portfolio')
plt.xlabel('Time Period')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(loc='upper left', fontsize='small', ncol = 5)  # Place legend outside the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
# Calculate the running maximum
running_max = cumulative_returns.cummax()
# Calculate the drawdown
drawdown = (cumulative_returns - running_max) / running_max
for index, row in drawdown.iterrows():
    plt.plot(row, label=f'Target Return: {returns.iloc[index]["Target Returns"]:.3f}')

plt.title('Portfolio Drawdown for Each Target Return Portfolio')
plt.xlabel('Time Period')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.ylabel('Drawdown')
plt.legend(loc='lower left', ncol = 2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot the distribution of returns for all rows
plt.figure(figsize=(14, 8))

for index in range(len(returns)):
    row_data = returns.iloc[index, 1:]  # Exclude 'Target Returns' column
    plt.hist(row_data, bins=15, alpha=0.5, label=f'Target Return: {returns.iloc[index, 0]:.3f}', edgecolor='k')

plt.title('Distribution of Returns for All Target Return Portfolios')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
